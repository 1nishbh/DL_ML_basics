```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(0, 100)]
```

```python
y = x
```

```python
y_natlog = np.log(x)
y_log10 = np.log10(x)
y_log2 = np.log2(x)
```

```python
print(y_log10[1:10], y_natlog[1:5], y_log2[1:5])
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y, label = 'y=x')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Linear Plot')
plt.text(40, y[40], "y = x", size= 11, horizontalalignment='right')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
#plt.plot(x, y, label='y = x')
plt.plot(x, y_natlog)
plt.plot(x, y_log10)
plt.plot(x, y_log2)

# Add labels on the lines
plt.text(20, y_natlog[20], 'ln(x)', fontsize=10, verticalalignment='top')
plt.text(20, y_log10[20], 'log10(x)', fontsize=10, verticalalignment='top')
plt.text(20, y_log2[20], 'log2(x)', fontsize=10, verticalalignment='top')

plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Logarithmic Plots')
plt.grid(True)
plt.show()
```

The sigmoid function transforms the continuous real number into a range of ( 0 , 1 ) , so that the input value of the next layer is within a fixed range and the weight is more stable.

```python
# sigmoid function
x = np.arange(-50, 50)
y = x
y_sigmoid = 1 / (1 + np.exp(-x))
```

```python
print(y_sigmoid[0:5], y_sigmoid[-10:])
```

```python
plt.figure(figsize=(10,6))
#plt.plot(x, y, label = 'y=x')
plt.plot(x, y_sigmoid, label = "sigmoid transformation")
plt.legend()
plt.show()
```

## Rectified Linear Unit
- commonly used as activation function in Deep learning
- returns 0 for x < 0 and x for x >= 0

```python
def relu(x):
    return 0 if x < 0 else x
```

```python
x = np.arange(-50, 50)
y = x
y_relu = [relu(x) for x in x]
```

```python
plt.plot(x, y, label = "y = x")
plt.plot(x, y_relu, label = "y = relu(x)")
plt.legend()
plt.xlabel("input")
plt.ylabel("output")
plt.show()
```
